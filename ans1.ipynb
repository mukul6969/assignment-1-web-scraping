```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
titles=soup.find('h1',class_="firstHeading")
titles
```

```python
titles=soup.find_all('div',class_="mw-parser-output")
titles
```

```python
titles=soup.find('h2',class_="mp-h2")
titles
```

```python
first_company=soup.find('span',class_="mw-headline")
first_company
```
